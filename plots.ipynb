```python
import matplotlib
import pandas as pd
```

```python
day = pd.read_csv("day_analysis.csv")
day.plot(y='center', xlabel="Frame #", ylabel="Center (Day)")
day['percentage'] = (day['max_brightness_count'] / day['total_count']) * 100
day.plot(y='percentage', xlabel="Frame #", ylabel="Center (Day)")

day['percentage2'] = (day['min_brightness_count'] / day['total_count']) * 100
day.plot(y='percentage2', xlabel="Frame #", ylabel="Percentage brightness (Day)")
```

```python
day = pd.read_csv("night_analysis.csv")
day.plot(y='center', xlabel="Frame #", ylabel="Center (Night)")
day['percentage'] = (day['max_brightness_count'] / day['total_count']) * 100
day.plot(y='percentage', xlabel="Frame #", ylabel="Center (Day)")


day['percentage2'] = (day['min_brightness_count'] / day['total_count']) * 100
day.plot(y='percentage2', xlabel="Frame #", ylabel="Percentage brightness (Night)")
```
